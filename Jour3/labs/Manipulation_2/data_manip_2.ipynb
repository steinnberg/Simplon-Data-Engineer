* `df[['IMDb', 'Netflix']]` : On garde seulement deux colonnes du DataFrame :
    - IMDb : par exemple, la note du film.
    - Netflix : une valeur indiquant s’il est dispo sur Netflix (souvent 0 = non, 1 = oui).
* `.groupby('Netflix')` : On regroupe les lignes du tableau selon la colonne Netflix. Donc, on crée 2 groupes :

`df.groupby('Netflix')`

➡️ On regroupe les lignes du DataFrame selon la colonne Netflix.

- Chaque groupe correspond à une valeur différente de Netflix (par exemple 0 = non, 1 = oui).

    - On crée donc deux sous-ensembles de données :

        - Les films non disponibles sur Netflix
        - Les films disponibles sur Netflix

```python
df.groupby('Netflix').describe().T
```

```python
# #which serie is the best in each categorie
# test = df[['IMDb', 'Age']].groupby('Age')
# #test
# new_series = pd.Series()

# for i ,j in test: #i is the group, j is the dataframe
#     new_series = new_series.append(j.IMDb.rank(ascending= False))
# new_series.sort_index()

# Groupby 'Age'
grouped = df[['IMDb', 'Age']].groupby('Age')

# Liste pour stocker toutes les sous-séries rankées
ranks = []

for age_group, group_df in grouped:
    # Calcule le rang dans le groupe
    rank = group_df['IMDb'].rank(ascending=False)
    ranks.append(rank)

# Concatène proprement toutes les Series
new_series = pd.concat(ranks)

# Trie pour remettre dans l'ordre d'origine
new_series = new_series.sort_index()

print(new_series.head())
```
